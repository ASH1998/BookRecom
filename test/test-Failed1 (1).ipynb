```python
counter = 0
for i in samplereview:
    try:
        review0 = pd.read_feather(i)
        for j in tqdm.tqdm(sampleindex):
    #         print(i, '->>', j)
            try:
                index0 = pd.read_feather(j)
                index0['bookid'] = index0['book_id']
                res = pd.merge(index0, review0, on=['bookid'])
                res = res[['isbn', 'text_reviews_count', 'country_code', 'language_code', 'average_rating',
             'ratings_count', 'title', 'title_without_series', 'bookid', 'rating', 'reviewtext',  'description',
            'nvotes', 'image_url']]
                res.to_feather('resultantFolder/resultSet'+str(counter)+'.ftr')
                counter+=1
            except:
                continue
    except:
         continue
```
